```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Array with Numpy

a = np.array([4,7,68,10,9])
print(type(a))
print(a.shape)
print(a[0])
a[0]
print(a)
```

```python
b = np.array([[5, 6, 7],[6, 2, 4]])
print(b.shape)
#printing the entire matrix
print(b)

#printing the position (1,1).
print(b[b])
```

```python
#Plotting arrays using Matplotlib

list_int = np.array([5,7,8,10,13,25])
plt.plot(list_int)
plt.ylabel("integer numbers")
plt.xlabel( "x axis")
plt.show()
```

```python
#Formatting the style of yout plot
plt.plot([2,3,6,7], [10,7,4,1], 'r*')
plt.show()
```

```python
t = np.arange(0.,5.,0.2)

#red dashes, blue square and green triangles
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)
plt.figure()
plt.subplot(211)
plt.plot(t1,f(t1),'bo', t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
#Create an Empty DataFrame

df = pd.DataFrame()
print(type(df))
```

```python
#Create a DataFrame from a array
data = [1,4,7,9,10]
df= pd.DataFrame
```

```python
data = [['Alex',10],['Bod',12],['Clarke',13]]
df = pd.DataFrame(data,columns=['Name','Age'])
print(df)
```

```python
#Setting the datatype on your DataFrame

df_float = pd.DataFrame(data,columns=['Name','Age'], dtype=float)
print(df_float)
```

```python
#Create a DataFrame from a list

data_list = {'Name':['Tom','Jack','Steve','Carlos'], 'Age':[12,23,16,9]}

df_list = pd.DataFrame(data_list, index=['first','second','third','fourth'])
print(df_list)
```

```python
#Create a dataframe from a dict of series

dic = {'one': pd.Series([1,4,6],index=['a','b','c']),
       'two': pd.Series([1,2,3,4],index=['a','b','c','d'])}
df = pd.DataFrame(dic)
print(df)
```
